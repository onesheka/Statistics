# STATISTICS - Scales, Histograms and measuring values

### 1. DISCRETE AND CONTINUOUS SCALES

```python
day_number = [1,2,3,4,5,6]
chocolates_eaten_per_day = [2,4,6,3,2,3] ## discrete values
distance_travelled_per_day = [2.2,1.1,3.4,5.6,1.1,2.2] ## continious values

import matplotlib.pyplot as plt

plt.plot(day_number,chocolates_eaten_per_day)
plt.show()

plt.plot(day_number,distance_travelled_per_day)
plt.show()
```

### 2. SCALE STARTING POINTS

Some scales don't have 0 as their starting points, their absolute zeroes are something else. Like degrees in Farenheit. 
Absolute zero is at -456.67 degrees

```python
ft_degrees = [35, 54, 72, 89] ##if we need to convert the code so that absolute zero is at 0 only
for i in range(0,len(ft_degrees)):
    ft_degrees[i] = round(ft_degrees[i] + 456.67,2)
print(ft_degrees) ##we sometimes convert it bcoz it helps in finding ratios and relating.
```

### 3. ORDINAL SCALES

```python
survey_how_much_do_you_worry = ['a lot','not much','average','a lot','a lot','not much','a lot']
set_survey = set(survey_how_much_do_you_worry)
list_survey = list(set_survey)
print(list_survey)
responses = []
import numpy as np
for i in range(0,len(survey_how_much_do_you_worry)):
    responses.append(list_survey.index(survey_how_much_do_you_worry[i]))

print(responses)
print(round(np.mean(responses),2))

## as mean is close to 2, according to the survey maximum people worry a lot :(
```

### 5. CATEGORICAL SCALES

```python
gender = ['male','female','female','male','female','male']
marks = [86,90,92,55,80,70]

female_marks =[]
male_marks = []
for i in range(0,len(gender)):
    if(gender[i]=="female"):
        female_marks.append(marks[i])
    else:
        male_marks.append(marks[i])
        
print("Average Female marks: ",round(np.mean(female_marks),2))
print("Average Male marks: ",round(np.mean(male_marks),2))
```

### 8. DATA SKEW

```python
## show the distrubutional characteristics of the data
scores_1 = [56,78,89,92,93,95,96]
scores_2 = [45,56,67,56,78,65,95]
scores_3 = [35,45,55,65,75,85,70]

plt.hist(scores_1,bins=4)
plt.show()  ##negative skew

plt.hist(scores_2,bins=4)
plt.show()  ##positive skew

plt.hist(scores_3,bins=4)
plt.show()  ##no skew
```

```python
from scipy.stats import skew
positive_skew = skew(scores_2)
negative_skew = skew(scores_1)
no_skew = skew(scores_3)

print(positive_skew)
print(negative_skew)
print(no_skew) ##very slightly towards the left side
```

### 9. KURTOSIS

### 11. MEAN

```python
scores = [55,65,78,87,95]
mean_score = np.mean(scores)

plt.hist(scores)
plt.axvline(mean_score,color="r")
plt.show()
```

### 12.MEDIAN

```python
## midpoint of the sorted array
## less sensitive to very large or very small values
median_score = np.median(scores)
plt.hist(scores)
plt.axvline(median_score,color="r")
plt.show()
```
