# COMPARISON OF FANDANGO AND METACRITIC MOVIE REVIEWS
Continuation of the project....Comapring these 2 particular site reviews

```python
import pandas as pd
movie_reviews = pd.read_csv("fandango_scores.csv")
movie_reviews
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(movie_reviews['Fandango_Stars'],[0,1,2,3,4,5])
plt.show()
plt.hist(movie_reviews['Metacritic_norm_round'],[0,1,2,3,4,5])
plt.show()
```

Here, we can observe that Fnadango scores start their ratings from 3 stars and maximum values are around 4-5 stars.
Whereas, talking about the Metacritic star ratings, it's quite spread out and maximum values fall in 3-4 range.

```python
import numpy as np
fandango_mean = np.mean(movie_reviews['Fandango_Stars'])
metacritic_mean = np.mean(movie_reviews['Metacritic_norm_round'])
fandango_median = np.median(movie_reviews['Fandango_Stars'])
metacritic_median = np.median(movie_reviews['Metacritic_norm_round'])
fandango_std = np.std(movie_reviews['Fandango_Stars'])
metacritic_std = np.std(movie_reviews['Metacritic_norm_round'])

print("Fandango average rating:",fandango_mean)
print("Metacritic average rating:",metacritic_mean)
print("---------------------------------------")
print("Fandango Median:",fandango_median)
print("Metacritic Median:",metacritic_median)
print("---------------------------------------")
print("Fandango Standard Deviation:",fandango_std)
print("Metacritic Standard Deviation:",metacritic_std)


```

Here we can observe that Fandango has way higher average rating than Metacritic.
We can also observe that Median of Metacritic is more than it's average. The mean is usually lower than median when there are few small values. As there are comparitively less movies which have low stars(<2), so the median is higher than the average. <br>
Also, the standard deviation of Fandango is low as compared to that of Metacritic. This is because the range of Fandango is quite low as compared to that of Metacrtic.<br>
Fandango rates maximum movies on it's site quite high.

```python
plt.scatter(movie_reviews['Fandango_Stars'],movie_reviews['Metacritic_norm_round'],color="g")
plt.show()
```

```python
movie_reviews['diff_mov'] = movie_reviews['Fandango_Stars'] - movie_reviews['Metacritic_norm_round']
movie_reviews['diff_mov'] = abs(movie_reviews['diff_mov'])
movie_reviews.sort_values('diff_mov',ascending=False)[0:5]
```

```python
from scipy.stats import pearsonr
from scipy.stats import linregress
x = movie_reviews['Fandango_Stars']
y = movie_reviews['Metacritic_norm_round']
r_value,p_value = pearsonr(x,y)
slope,intercept,r_value,p_value,stderr = linregress(x,y)

pred_y = slope*3  + intercept
print("r between Fandango_Stars and Metacritic_norm_round:",r_value)
print("Slope:",slope)
print("Intercept:",intercept)
print("If Fandango would have a rating of 3, then Fandango rating can be predicted as:",pred_y)
```

Here, we can observe that Fandango has higher ratings for maximum movies as compared to Metacritic movie reviews

```python
pred_1 = slope*1 + intercept
pred_5 = slope*5 + intercept
plt.scatter(movie_reviews['Fandango_Stars'],movie_reviews['Metacritic_norm_round'])
plt.plot([1,5],[pred_1,pred_5],color="r")
plt.xlim(1,5)
plt.show()
```
